```python
df_test.shape
```

```python
def plot_design():
    plt.xlabel('')
    plt.ylabel('')
    plt.yticks(fontsize=13, color='black')
    plt.box(False)
    plt.title(i[1], fontsize=20, color='black')
    plt.tight_layout(pad=5.0)
    plt.grid(b=None)
```

```python
# Selecting categorical variables
categ = df_test.select_dtypes(include=object).columns

# Visualization
fig, ax = plt.subplots(figsize =(50, 100))
fig.patch.set_facecolor('white')
mpl.rcParams['font.family'] = 'sans-serif'
mpl.rcParams['font.size'] = 20

colors = sns.color_palette("Paired").as_hex()

# Using for loop to create plot for each column
for i in (enumerate(categ)):
    plt.subplot(22, 1, i[0]+1)
    
    if df_test[i[1]].value_counts().count() > 2:
        ax = sns.countplot(y = i[1], data = df_test, order=df_test[i[1]].value_counts().index, palette=colors)
        pct = df_test[i[1]].value_counts(ascending=False, normalize=True).values * 100
        ax.bar_label(container=ax.containers[0], labels=list(map('{:.2f}%'.format,pct)), padding=3, size=12, color='black')
        ax.grid(False)
        ax.grid(b=None)
        ax.xaxis.set_ticks_position('none')
        ax.yaxis.set_ticks_position('none')
        ax.set_xticklabels('')
        plot_design()
        
    else:
        _, texts, pcts = plt.pie(
            df_test[i[1]].value_counts(), 
            labels=df[i[1]].value_counts().index, 
            colors= ['#BF2F38', '#2F2F2F'],
            autopct='%1.1f%%', 
            wedgeprops={'linewidth': 1.0, 'edgecolor': 'white'},
            startangle=50)
        for pcts in pcts:
            pcts.set_color('white')
            plt.title(i[1], fontsize=20, color='black')
```

## Comparison For Top 3 & Bottom 3 Postcode Sector

I will compare each features of top 3 and bottom 3 postcode sector together to see which features are important for the performance.

#### Age_Band

The age of the customer is irrelevant to the campaign performance as the distribution of the age band for top 3 & bottom 3 postcode sector are similar.

#### Marital_Status
The marital status of the customer is irrelevant to the campaign performance as the distribution of the marital status for top 3 & bottom 3 postcode sector are similar.

#### Income_Band
For income band, the data shows that customer with lower income band tends to be more willing to purchase insurance. 

From the data of the best performing top 3 postcode sector, it shows there are 71.39% of customers have the lowest income band of £0-£9,999 and the second lowest income band of £10,000-£19,999. 

Meanwhile, from the data of the worst performing bottom 3 postcode sector, it shows the lowest income band of £0-£9,999 and the second lowest income band of £10,000-£19,999 only account for 48.53% and customers in the bottom 3 postcode sector mainly have a higher income band. 

One of the possible reasons for this phenomenon is because customers with higher income band would think that their salary and savings can support themselves when accident happens while customers with lower income band would think that insurance can protect them in bad days as they do have huge amount of savings due to their low and unstable incomes. 

Thus, it shows that customers with lower income band will be more likely to purchase insurance.

#### Social_Eco_Class
For Socio-economic classification band, the data shows that customer with lower social grade tends to be more willing to purchase insurance.

From the data of the best performing top 3 postcode sector, it shows there are 60.76% of customers have the lowest Socio-economic classification band of E and the second lowest Socio-economic classification band of D. 

Meanwhile, from the data of the worst performing bottom 3 postcode sector, it shows the lowest Socio-economic classification band of E and the second lowest Socio-economic classification band of D only take up 23.02% and customers in the bottom 3 postcode sector primarily have a higher social grade. 

One of the possible reasons for this phenomenon is because customers with lower social grade are self-employ, causal workers or unemployed which may not have a comprehensive benefit package to cover their insurance needs or even they do not have any insurance provided, thus they may tend to be more sensitive to the information of insurance from a door drop campaign. 

Thus, it shows that customers with lower social grade will be more likely to purchase insurance.

#### Occupation
After looking into the distinct occupations of the customers in the best performing top 3 postcode sector and the worst performing bottom 3 postcode sector, it shows that the occupations of customers are align with the socio-economic classification band mentioned above for the top 3 and bottom 3 postcode sector representatively and it further proved the customer with lower social grade tends to be more willing to purchase insurance.

#### Affluence_Rank
The affluence rank is from Grade 1, the lowest, to Grade 17, the highest. 

For affluence rank band, the data shows that customer with lower affluence rank tends to be more willing to purchase insurance.

In general, customers in the best performing top 3 postcode sector has a lower grade while the worst performing bottom 3 postcode sector has a higher grade. 

For the best performing top 3 postcode sector, there are only 11.9% of customers who are grade 8 or above.

For the worst performing bottom 3 postcode sector, there are 44.53% of customers who are grade 8 or above.

Thus, it shows that customers with lower affluence rank are more likely willing to purchase insurance.
